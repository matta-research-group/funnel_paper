```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import molplotly
import matplotlib.pyplot as plt
```

### Load in the dataframe

```python
data_df = pd.read_csv('../DataFrames/eg_plan_df.csv')
```

```python
data_df
```

### Set the desried colourpalette

```python
blind = sns.color_palette("colorblind")
```

```python
colors = [blind.as_hex()[0], blind.as_hex()[3], blind.as_hex()[4], blind.as_hex()[8], blind.as_hex()[2]]
```

```python
fig = plt.figure(figsize=(10, 8))
axes1 = fig.add_axes([0.1, 0.1, 1, 1])

h = sns.jointplot(ax=axes1, data=data_df, x='Planarity', y='HOMO-LUMO Energy Gap /eV', hue='Linker Type', palette=colors, s=30)
h.ax_joint.set_xlabel('Planarity = |cos(Î¸)|')
h.ax_joint.set_ylabel('$E_g$ (eV)')
h.ax_joint.legend(title='', bbox_to_anchor=(0.01, 0.28), loc='upper left')
h.ax_joint.set_xlim(0, 1.05)
h.ax_joint.set_ylim(1.5, 4.1)

width_in_inches = 17.1 / 2.54
# Get the current figure size in inches
fig = plt.gcf()  # Get the current figure
fig.set_size_inches(width_in_inches, fig.get_size_inches()[1])

plt.savefig('plot_pngs/planarity_eg_plot', bbox_inches='tight', dpi=1400)
```
