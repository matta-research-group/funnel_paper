For a demonstration of a interactive plot for planarity and energy gap


```python
import pandas as pd
import plotly.express as px
import molplotly

combined_df = pd.read_csv('../../DataFrames/eg_plan_df.csv')

fig = px.scatter(combined_df, x="Planarity", y="HOMO-LUMO Energy Gap /eV")

app = molplotly.add_molecules(fig=fig,
                            df=combined_df,
                            smiles_col=['SMILES'],
                                  caption_transform={'Planarity': lambda x: f"{x:.3f}",
                                                     'HOMO-LUMO Energy Gap /eV': lambda x: f"{x:.3f}"})

app.run(mode='mode', port=8100, height=1000)
```

For demonstration of plotting the reorganisation energy


```python
import pandas as pd
import plotly.express as px
import molplotly

combined_reorg_df = pd.read_csv('../../DataFrames/reorganisation_dyads_df.csv')

fig_reorg = px.scatter(combined_reorg_df, x="位-Cation /meV", y="位-Anion /meV")

app_reog = molplotly.add_molecules(fig=fig_reorg,
                            df=combined_reorg_df,
                            smiles_col=['SMILES'],
                                  caption_transform={'Planarity': lambda x: f"{x:.3f}",
                                                     'HOMO-LUMO Energy Gap /eV': lambda x: f"{x:.3f}",
                                                     'IP': lambda x: f"{x:.3f}",
                                                     'EA': lambda x: f"{x:.3f}",
                                                     '位-Cation /meV': lambda x: f"{x:.3f}",
                                                     '位-Anion /meV': lambda x: f"{x:.3f}"})

app_reog.run(mode='mode', port=9300, height=1000)
```
